```python
fish_data = pd.read_csv("data/DatasetClusters/fishes/fish02/fish_02_mts.csv")
fish_data_s1 = pd.read_csv("data/DatasetClusters/fishes/fish02/s1/s1_mts.csv")
fish_data_s2 = pd.read_csv("data/DatasetClusters/fishes/fish02/s2/s2_mts.csv")
```

```python
red_columns = [col for col in fish_data.columns if 'red' in col]
green_columns = [col for col in fish_data.columns if 'green' in col]
# Creating separate DataFrames for 'red' and 'green' columns
df_red = fish_data[red_columns]
df_green = fish_data[green_columns]
len(df_red)
```

```python
import re

# Function to extract cluster numbers and sort columns based on that
def sorted_columns_by_cluster(df, color):
    pattern = re.compile(rf"cluster_(\d+)_({color})")
    sorted_columns = sorted(df.columns, key=lambda x: int(pattern.search(x).group(1)))
    return sorted_columns

# Extract and sort columns for red and green based on cluster number
red_columns_sorted = sorted_columns_by_cluster(df_red, 'red')
green_columns_sorted = sorted_columns_by_cluster(df_green, 'green')
```

```python
import itertools
red_columns = df_red.columns
green_columns = df_green.columns

# Generate all possible pairs of red and green columns
all_pairs = list(itertools.product(red_columns, green_columns))

len(all_pairs)
```

```python
pairs_dic = {}
for i in range(len(all_pairs)):
    pairs_dic[i] = all_pairs[i]
```

```python
preprocessing_pipeline = Pipeline([
    ('scaler1', StandardScaler()),
])

```

```python
paired_df = pd.DataFrame()

for red_col, green_col in all_pairs:
    # Assuming each pair of columns should be concatenated side by side
    paired_data = pd.concat([df_red[red_col], df_green[green_col]], axis=1)
    
    # Optionally rename columns to reflect they are paired data
    paired_data.columns = [f'{red_col}_paired', f'{green_col}_paired']
    
    # Append the paired data to the initialized DataFrame
    paired_df = pd.concat([paired_df, paired_data], axis=1)
```

```python
data_preprocessed = preprocessing_pipeline.fit_transform(paired_df)
df_preprocessed = pd.DataFrame(data_preprocessed, columns=paired_df.columns, index=paired_df.index)
```

```python
fish_04_mts = pd.read_csv("data/DatasetClusters/fishes/fish04/fish_04_mts.csv")
fish_05_mts = pd.read_csv("data/DatasetClusters/fishes/fish05/fish_05_mts.csv")
fish_06_mts = pd.read_csv("data/DatasetClusters/fishes/fish06/fish_06_mts.csv")
fish_07_mts = pd.read_csv("data/DatasetClusters/fishes/fish07/fish_07_mts.csv")

```

```python
# Initialize a new empty DataFrame with the required length (assuming all have the same length)
new_df_length = len(fish_04_mts)
new_df = pd.DataFrame(index=range(new_df_length))

# List of DataFrames for easier management
dfs = [fish_04_mts, fish_05_mts, fish_06_mts, fish_07_mts]
fish_names = ["fish_04", "fish_05", "fish_06", "fish_07"]  # Names for prefix

def select_two_columns(df, prefix):
    # Randomly select two columns
    selected_columns = np.random.choice(df.columns, 2, replace=False)
    # Create new column names with prefix
    new_column_names = {selected_columns[0]: f"{prefix}_{selected_columns[0]}", 
                        selected_columns[1]: f"{prefix}_{selected_columns[1]}"}
    # Return the two columns with new names
    return df[selected_columns].rename(columns=new_column_names)

# Iterate to fill the new DataFrame
for _ in range(384):  # You need 768/2 iterations
    # Select a random index to pick DataFrame and corresponding prefix
    idx = np.random.randint(0, len(dfs))
    random_df = dfs[idx]
    fish_prefix = fish_names[idx]
    # Get two columns from the selected DataFrame
    two_columns = select_two_columns(random_df, fish_prefix)
    # Concatenate these columns into the new DataFrame
    new_df = pd.concat([new_df, two_columns], axis=1)

# Now new_df has 768 columns as required
```
